```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3','K4','K5'],
                         'A': ['A0', 'A1', 'A2', 'A3','A4','A5'],
                         'B': ['B0', 'B1', 'B2', 'B3','B4','B5']})

right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3','K6','K5'],
                          'C': ['C0', 'C1', 'C2', 'C3','C6','C7'],
                         'D': ['D0', 'D1', 'D2', 'D3','D6','D7']})
```

```python
left
```

```python
right
```

```python
left.merge(right,on='key')
```

```python
left.merge(right,on='key',how='right')
# keeps the value of own table and writes other 
```

```python
left.merge(right,on='key',how='left')
```
