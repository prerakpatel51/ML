```python
import pandas as pd 
import numpy as np 
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X,y=make_regression(n_samples=4,n_features=1,n_targets=1,n_informative=1,noise=80,random_state=13)

# X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X,y)
plt.show()
```

```python
model=LinearRegression()
```

```python

model.fit(X,y)
```

```python
model.coef_  # slope m
```

```python
model.intercept_  # b
```

```python
y_ols=model.predict(X)
```

```python
plt.scatter(X,y)
plt.plot(X,y_ols)
```

```python
# let assume that the slope is constant m=78.35
# and the starting value of the b is b=0
```

```python
y_pred=((78.35*X)+0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,y_pred,label='b=0',c='r')
plt.plot(X,y_ols,label='ols')
plt.legend()
plt.show()
# prediction for line where m=78.35 and assuming b=0 and the plotiing it with the regression  line
```

```python
# 1St iteration
m=78.35
b=0
loss_slope=-2*np.sum(y - m * X.ravel() - b)
loss_slope
```

```python
lr=0.1
step_size=loss_slope*lr
step_size
```

```python
b= b - step_size
b
# this is the new intercept
```

```python
y_pred1=((78.35*X)+b).reshape(4)
```

```python

plt.scatter(X,y)
plt.plot(X,y_pred,label='b=0',c='r')
plt.plot(X,y_pred1,label='b=20.92',c='purple')

plt.plot(X,y_ols,label='ols')
plt.legend()
plt.show()
# prediction for line where m=78.35 and assuming b=0 and the plotiing it with the regression  line
```

```python
# 2nd iteration
loss_slope=-2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
step_size=lr*loss_slope
step_size
```

```python
b=b-step_size
b
```

```python
y_pred2=((m*X)+b).reshape(4)
plt.scatter(X,y)
plt.plot(X,y_pred,label='b=0',c='r')
plt.plot(X,y_pred1,label='b=20.92',c='purple')
plt.plot(X,y_pred2,label='b=25.11',c='green')

plt.plot(X,y_ols,label='ols')
plt.legend()
plt.show()
```

```python
# 3rd iteration
loss_slope=-2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
step_size=lr*loss_slope
step_size
```

```python
b=b-step_size
b
```

```python
y_pred3=((m*X)+b).reshape(4)
plt.figure(figsize=(15,15))
plt.scatter(X,y)
plt.plot(X,y_pred,label='b=0',c='r')
plt.plot(X,y_pred1,label='b=20.92',c='purple')
plt.plot(X,y_pred2,label='b=25.11',c='green')
plt.plot(X,y_pred3,label='b=25.95',c='black')

plt.plot(X,y_ols,label='ols')
plt.legend()
plt.show()
```

```python
# lets do this in loops
plt.figure(figsize=(15,15))

b=0
m=78.35
lr=0.01
epoch=10
for i in range(epoch):
    loss_slope=-2*np.sum(y-m*X.ravel()-b)
    step_size=lr*loss_slope
    b=b-step_size
    print("loss_slope",loss_slope)
    print("step_size",step_size)
    print("b",b)
    y_pred= m*X+b
    plt.plot(X,y_pred) 
plt.scatter(X,y)
plt.show()
```
