```python
import pandas as pd 
import requests
```

```python
request=requests.get("https://api.themoviedb.org/3/movie/top_rated?api_key=084a1dc314df44c153a8ee7c45817106&language=en-US&page=1")
```

```python
df=pd.DataFrame(request.json()['results'])[['id','original_title','popularity','vote_count','genre_ids']]
```

```python
df.head()
```

```python
df=pd.DataFrame()
```

```python
for i in range(1, 453):
    request = requests.get(f"https://api.themoviedb.org/3/movie/top_rated?api_key=084a1dc314df44c153a8ee7c45817106&language=en-US&page={i}")
    tdf = pd.DataFrame(request.json()['results'])[['id', 'original_title', 'popularity', 'vote_count', 'genre_ids']]
    df = pd.concat([df, tdf], ignore_index=True)  # Use pd.concat instead of append
```

```python
df
```

```python
df.to_csv('movies.csv')
```

```python
request=requests.get(f"https://www.opendata.nhs.scot/api/3/action/datastore_search?resource_id=23b3bbf7-7a37-4f86-974b-6360d6748e08")
df=pd.DataFrame(request.json()['result']['records'])

for i in range(1,178):
    request=requests.get(f"https://www.opendata.nhs.scot/api/3/action/datastore_search?offset={i*100}&resource_id=23b3bbf7-7a37-4f86-974b-6360d6748e08")
    tempdf=pd.DataFrame(request.json()['result']['records'])
    df=pd.concat([df,tempdf],ignore_index=True)
```
